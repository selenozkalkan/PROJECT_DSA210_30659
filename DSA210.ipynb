```python
# Convert the 'ts' column to datetime
df_spotify['timestamp'] = pd.to_datetime(df_spotify['ts'])

# Select relevant columns for analysis
df_spotify = df_spotify[[
    'timestamp',
    'ms_played',
    'master_metadata_track_name',
    'master_metadata_album_artist_name',
    'master_metadata_album_album_name',
    'reason_start',
    'reason_end'
]]

# Rename columns for clarity
df_spotify.rename(columns={
    'master_metadata_track_name': 'track_name',
    'master_metadata_album_artist_name': 'artist_name',
    'master_metadata_album_album_name': 'album_name'
}, inplace=True)

# Preview the cleaned data
print("Cleaned Spotify Data:")
print(df_spotify.head())

```

```python
# Remove timezone from 'timestamp' in df_spotify
df_spotify['timestamp'] = df_spotify['timestamp'].dt.tz_localize(None)

# Ensure 'start_date' and 'end_date' in df_exam_periods are timezone-naive
df_exam_periods['start_date'] = df_exam_periods['start_date'].dt.tz_localize(None)
df_exam_periods['end_date'] = df_exam_periods['end_date'].dt.tz_localize(None)
```

```python
# Add an 'exam_period' column
df_spotify.loc[:, 'exam_period'] = None
```

```python
# Assign exam periods to Spotify records
for _, row in df_exam_periods.iterrows():
    mask = (df_spotify['timestamp'] >= row['start_date']) & (df_spotify['timestamp'] <= row['end_date'])
    df_spotify.loc[mask, 'exam_period'] = row['label']

# Preview the labeled Spotify data
print("Spotify Data Labeled with Exam Periods:")
print(df_spotify[['timestamp', 'track_name', 'exam_period']].head())
```

Hypothesis Test:


```python
# Create a list of exam days in 2024
exam_days_2024 = set()
for _, row in df_exam_periods.iterrows():
    if row['start_date'].year == 2024:
        exam_days_2024.update(pd.date_range(row['start_date'], row['end_date']))

# Separate daily playback data into exam and non-exam days
daily_playback_exam_2024 = daily_playback_2024[daily_playback_2024.index.isin(exam_days_2024)]
daily_playback_non_exam_2024 = daily_playback_2024[~daily_playback_2024.index.isin(exam_days_2024)]

# Calculate daily averages as independent samples
exam_daily_averages = daily_playback_exam_2024.values
non_exam_daily_averages = daily_playback_non_exam_2024.values

print("Daily Averages for Exam Periods:", exam_daily_averages[:10])
print("Daily Averages for Non-Exam Periods:", non_exam_daily_averages[:10])
```

```python
from scipy.stats import ttest_ind

# Perform Welch's t-test
t_stat, p_value_two_sided = ttest_ind(
    exam_daily_averages,
    non_exam_daily_averages,
    equal_var=False  # Welch's t-test
)

# Convert two-sided p-value to one-sided
p_value_one_sided = p_value_two_sided / 2 if t_stat > 0 else 1 - (p_value_two_sided / 2)

# Print results
print("=== Hypothesis Test Results ===")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value (One-Sided): {p_value_one_sided:.4f}")

# Conclusion based on significance level
alpha = 0.05
if p_value_one_sided < alpha:
    print("Conclusion: Reject the null hypothesis. Playback is significantly higher during exam periods.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No evidence that playback is higher during exam periods.")
```
